```python
from IPython.display import clear_output
import time

board = [""] * 9
def print_board(board):
    print("\n")
    print(f" {board[0]} | {board[1]} | {board[2]}")
    print("---------")
    print(f" {board[3]} | {board[4]} | {board[5]}")
    print("---------")
    print(f" {board[6]} | {board[7]} | {board[8]}")
    print("\n")

current_player = "X"

def check_winner(board, current_player):
    wins = [
        [0, 1, 2], [3, 4, 5], [6, 7, 8],  # Reihen
        [0, 3, 6], [1, 4, 7], [2, 5, 8],  # Spalten
        [0, 4, 8], [2, 4, 6]              # Diagonalen
    ]
    return any(board[i] == board[j] == board[k] == current_player for i, j, k in wins)

while True:
    clear_output(wait=True)
    print_board(board)
    time.sleep(0.2)

    move = input(f"Spieler {current_player} enter your move (1-9): ")

    if move.lower() == "exit":
        print("Spiel wird beendet.")
        break

    if not move.isdigit() or not 1 <= int(move) <= 9:
        print("Ungültige Eingabe. Bitte eine Zahl zwischen 1 und 9 eingeben.")
        time.sleep(1.5)
        continue

    move = int(move) - 1
    if board[move] != "":
        print("Feld bereits belegt. Bitte ein anderes Feld wählen.")
        time.sleep(1.5)
        continue

    board[move] = current_player

    if check_winner(board, current_player):
        clear_output(wait=True)
        print_board(board)
        print(f"Spieler {current_player} gewinnt!")
        break


    if "" not in board:
        clear_output(wait=True)
        print_board(board)
        print("Unentschieden!")
        break

    current_player = "O" if current_player == "X" else "X"

```
